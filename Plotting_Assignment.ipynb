```python
co2_emissions = np.array(data['co2'])

# Range definition
x_min, x_max = 0, 750
x_step = 70

bin_edges = np.arange(x_min, x_max + x_step, x_step)

# plt.figure(figsize=(10, 6))
plt.hist(co2_emissions, bins=bin_edges, edgecolor="white")
plt.xticks(np.arange(x_min, x_max + x_step, x_step))
plt.title('Carbon Dioxide Distribution')
plt.xlabel('CO2')
plt.ylabel('Vehicle Frequency')
plt.show()
```

```python
# YOUR CODE HERE
cylinders = np.array(data['cylinders'])

# Range definition
x_step = 1
x_min, x_max = 0, int(max(cylinders)) + x_step

bin_edges = np.arange(x_min, x_max + x_step, x_step)
# print(bin_edges)

# plt.figure(figsize=(10, 6))
plt.hist(cylinders, bins=bin_edges, edgecolor="black")
plt.xticks(np.arange(x_min, x_max + x_step, x_step))
plt.title('ICE Cylinder Distribution')
plt.xlabel('Cylinders')
plt.ylabel('Vehicle Frequency')
plt.show()
```

```python
# YOUR CODE HERE
cylinders = np.array(data['displ'])

# Range definition
x_step = 1
x_min, x_max = 0, int(max(cylinders)) + (x_step * 2)

bin_edges = np.arange(x_min, x_max + x_step, x_step)

plt.hist(cylinders, bins=bin_edges, edgecolor="black")
plt.xticks(np.arange(x_min, x_max + x_step, x_step))
plt.title('ICE Displacement')
plt.xlabel('Displacement Size')
plt.ylabel('Vehicle Frequency')
plt.show()
```

```python
from matplotlib.colors import LinearSegmentedColormap

corr_matrix = data.corr(numeric_only=True)

colors = ["#C3553A", "#F2F0EF" ,"#408094"]
n_bins = 100

custom_cmap = LinearSegmentedColormap.from_list("custom", colors, N=n_bins)

sns.heatmap(corr_matrix, cmap=custom_cmap,  vmin=-1, vmax=1)

plt.xticks(rotation=45, ha='right')
plt.title('Correlation Matrix Heatmap', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

# Data Interpretation

## Carbon Dioxide Distribution

### Summary
This diagram shows the distribution of CO2 emmisions between vehicle models. The diagram shows that most vehicles in the dataset have CO2 emmisions that fall between 280 - 350, and the lowest number of vehicles between 0 - 70.

### Findings
  - The histogram leans slightly to the right meaning there are some vehicle with higher CO2 emmisions

  - There are very few vehicles with low emmisions (0 - 70) suggesting that as the emmisions tend towards zero there are fewer vehicles in the dataset

### Summary
The correlation heatmap helps map the degree of correlation between any two columns(features) in a data set. In this diagram it ranges from 1 to -1 where zero means no correlation at all.

### Findings
